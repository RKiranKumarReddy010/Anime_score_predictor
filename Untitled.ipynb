```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python

dr['episodes'] = dr['episodes'].fillna(0)
dr['members'] = dr['members'].fillna(0)
dr['rank'] = dr['rank'].fillna(0)
dr['popularity'] = dr['popularity'].fillna(0)
dr['Animation_score'] = dr['Animation_score'].fillna(0)
```

```python
from sklearn.model_selection import train_test_split
X = np.array(dr[["episodes","rank","popularity"]])
Y = np.array(dr[["Animation_score"]])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y,test_size=0.33)
```

```python
from sklearn import linear_model
reg = linear_model.Ridge(alpha = 0.5)
las = linear_model.Lasso(alpha = 0.5)
llars = linear_model.LassoLars( alpha = 0.1)
bay = linear_model.BayesianRidge()
```

```python

reg.fit(xtrain,ytrain)
las.fit(xtrain,ytrain)
llars.fit(xtrain,ytrain)
bay.fit(xtrain,ytrain)
```

```python
print("Anime Rank Predictor:")
c = input("Enter the name:")
a = float(input("Enter the number of episodes:"))
b = float(input("Enter the rank:"))
d = float(input("Enter the popularity:"))
f = np.array([[a,b,d]])
print("The Animation score will be:(RIDGE)",reg.predict(f))
print("The Animation score will be:(LASSO)",las.predict(f))
print("The Animation score will be:(LASSO_lars)",llars.predict(f))
print("The Animation score will be:(Bayesian)",bay.predict(f))
```
